```python
import os


EPOCHS = 1000
BATCH_SIZE = 8
ARCH = "flor"

IMG_SIZE = (128,32, 1)
DATA_ROOT_PATH = "../data"
IMAGES_PATH = os.path.join(DATA_ROOT_PATH, "images", "data")
IMAGES_PATH
```

```python
#read training data
train_data = None

with open( "train.txt", encoding="utf-8") as f:
  train_data = f.readlines()
train_data = [item.strip() for item in train_data]
```

```python
#read test data
test_data = None

with open( "test.txt", encoding="utf-8") as f:
  test_data = f.readlines()
test_data = [item.strip() for item in test_data]
```

```python
#read valid data
valid_data = None

with open( "val.txt", encoding="utf-8") as f:
  valid_data = f.readlines()
valid_data = [item.strip() for item in valid_data]
```

```python
valid_data
```

```python
#store all vocabulory words in vocab
vocab=[]
```

```python
#add all training data to vocab
for i in range(len(train_data)):
    curr_data = train_data[i].split()
    if curr_data[1] not in vocab:
        vocab.append(curr_data[1])
```

```python
#add all test data to vocab
for i in range(len(test_data)):
    curr_data = test_data[i].split()
    if curr_data[1] not in vocab:
        vocab.append(curr_data[1])
```

```python
#add all val data to vocab
for i in range(len(valid_data)):
    curr_data = valid_data[i].split()
    if curr_data[1] not in vocab:
        vocab.append(curr_data[1])
```

```python
#store vocabulory of STR dataset
file = open('VocabSTR.txt','w')
for item in vocab:
    file.write(item+"\n")
file.close()
```

```python
#read vocabulory of hindi HTR dataset
vocabHTR = None

with open( "/home/project/Desktop/shashank/ocr-hindi/data/hindi_vocab.txt", encoding="utf-8") as f:
  vocabHTR = f.readlines()
vocabHTR = [item.strip() for item in vocabHTR]
```

```python
#add new words from Hindi HTR dataset to vocabulory
for i in range(len(vocabHTR)):
    curr_data = vocabHTR[i]
    if curr_data not in vocab:
        vocab.append(curr_data)
```

```python
#store final vocab
file22 = open('FinalVocab.txt','w')
for item in vocab:
    file22.write(item+"\n")
file22.close()
```
